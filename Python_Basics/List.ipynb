```python
list1 = [4,3,1,-1,-5,10]

print(list1)
list1.sort()
# inplace sorting
print(list1)

sortedList = sorted(list1)
# Creates new sorted List
print(list1)
print(sortedList)

list2 = [0] * 5
print(list2)

new_list = sortedList+list2+list1
print(new_list)


```

```python
mylist = [1,2,3,4,5,6,7,8,9]

a = mylist[1:5]
print(a)

a = mylist[:5]
print(a)

a = mylist[1:]
print(a)

a = mylist[::1]
print(a)

# by default step index is 1
a = mylist[::2]
# step index is here 2, hence from start print all list with 2 jump
print(a)

a = mylist[2::]
print(a)
# default step index of 1, print all list from 2 idx with 1 jump

a = mylist[2::55]
print(a)
a = mylist[2::3]
print(a)


a = mylist[::-1]
print(a)
# reverse entire string

a = mylist[2::-1]
print(a)

a = mylist[2:100:2]
# Get elements from index 2 to 100(can give any range), skipping every other element
print(a)
```

```python
list_org = ["banana", "cherry", "apple"]

list_cpy = list_org
# Deep Copy, reference Value

list_cpy.append("lemon")
print(list_cpy)
print(list_org)

# shallow copy
list_c1 = list_org.copy()
list_c2 = list(list_org)
list_c3 = list_org[:1]
list_c1.append("Strawberry")
list_c2.append("Lychee")
list_c3.append("Pine Apple")
# only root level changes are allowed, if there is a nested list then deep copy is made

print(list_cpy)
print(list_org)
print(list_c1)
print(list_c2)
print(list_c3)


```
